```python
#PROBLEME
#alors je veux merger mon fichier "repos.csv", "psr.csv", "issues.csv" sauf que j'ai une erreur lorsque j'utilise 
#méthode merge

#doc
#https://www.digitalocean.com/community/tutorials/pandas-merge-two-dataframe#pandas-dataframe-merge-examples
repos = pd.read_csv('repos.csv')
prs = pd.read_csv('prs.csv')
repos_df = pd.DataFrame(repos)
prs_df = pd.DataFrame(prs)
prs_df = prs_df.rename(columns={'name': 'language'})

concat = pd.merge(prs_df, repos_df, how='inner')

langages_df = pd.merge(repos_df, prs_df, how="right", on=["language", "language"])

langages_df
```

```python
#calcul de la moyenne
moyenne = langages_df['num_repos'].rolling(300).mean()

langages_list = langages_df[langages_df['num_repos'] > moyenne];
langages_list
```

```python
Lg = langages_list[langages_list['year'] >= 2019];Lg
```

```python
Lg = langages_list[langages_list['count'] >= 100000];Lg
sbn.relplot(x="language",y="count",data=Lg,height=7)
```

```python
#regardons l'evolution des langages par rapport au temps
sbn.relplot(x='count', y='year', data=Lg, height=8)
```

```python
sbn.relplot(x='language', y='year', data=Lg, kind='line')
```

```python
sbn.set_theme(style="ticks", font_scale=1.25)
g = sbn.relplot(
    data=Lg,
    x="num_repos", y="count", hue="language",
    palette="crest", marker="x", s=100,
)
```

```python
sbn.relplot(x='year', y='num_repos', data=Lg, kind='line')
plt.show()
```

```python
corr = Lg.iloc[:,-15:].corr()
#utilisation de la methode corr() pou
f, ax = plt.subplots(figsize=(30,5))
sbn.heatmap(corr, annot=True, ax =ax)
plt.savefig('heatmap.png')
```
